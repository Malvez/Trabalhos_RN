```python
import tarfile
import os
import requests
from transformers import pipeline
```

```python
#Baixar e extrair a base de dados IMDB
url = 'https://ai.stanford.edu/~amaas/data/sentiment/aclImdb_v1.tar.gz'
response = requests.get(url, stream=True)
with open('aclImdb_v1.tar.gz', 'wb') as file:
    file.write(response.content)

with tarfile.open('aclImdb_v1.tar.gz', 'r:gz') as tar:
    tar.extractall()
```

```python
#Ler os arquivos de texto e armazenar os reviews
def read_reviews_from_dir(directory):
    reviews = []
    for filename in os.listdir(directory):
        if filename.endswith('.txt'):
            with open(os.path.join(directory, filename), 'r', encoding='utf-8') as file:
                reviews.append(file.read())
    return reviews

train_pos_reviews = read_reviews_from_dir('aclImdb/train/pos')
train_neg_reviews = read_reviews_from_dir('aclImdb/train/neg')

# Usando apenas uma amostra ou não dos dados
sample_reviews = train_pos_reviews[:100] + train_neg_reviews[:100]
#sample_reviews = train_pos_reviews + train_neg_reviews
```

```python
#Truncar os reviews para 512 tokens
max_length = 512

def truncate_review(review, max_length):
    return review[:max_length]

truncated_reviews = [truncate_review(review, max_length) for review in sample_reviews]

# Usar a classe pipeline da biblioteca transformers para análise de sentimento
sentiment_analysis = pipeline('sentiment-analysis')

# Analisar os sentimentos das amostras truncadas
results = sentiment_analysis(truncated_reviews)
```

```python
#Exibir alguns resultados

for i, (review, result) in enumerate(zip(truncated_reviews, results)):
    print(f'Review {i+1}: {review[:60]}...')
    print(f'Sentiment: {result["label"]}, Score: {result["score"]}\n')
```
